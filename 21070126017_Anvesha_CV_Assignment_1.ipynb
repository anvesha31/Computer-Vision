```python
# 1) Sampling and Quantization
```

```python
import cv2
import numpy as np
from google.colab import files
import matplotlib.pyplot as plt

# Function to perform sampling on an image
def image_sampling(image_path, factor):
    # Read the image
    image = cv2.imread(image_path)

    # Performing sampling
    sampled_image = image[::factor, ::factor, :]

    return sampled_image

# Function to perform quantization on an image
def image_quantization(image, levels):
    # Converting the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Performing quantization
    quantized_image = np.floor(gray_image / (256 / levels)) * (256 / levels)

    # Converting back to 3 channels for display
    quantized_image_rgb = cv2.cvtColor(quantized_image.astype(np.uint8), cv2.COLOR_GRAY2RGB)

    return quantized_image_rgb

# Function to visualize the original, sampled, and quantized images
def visualize_sampling_quantization(image_path, sampling_factor, quantization_levels):
    # Reading the original image
    original_image = cv2.imread(image_path)

    # Performing sampling
    sampled_image = image_sampling(image_path, sampling_factor)

    # Performing quantization
    quantized_image = image_quantization(original_image, quantization_levels)

    # Displaying the original, sampled, and quantized images using matplotlib
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')


    plt.subplot(1, 3, 2)
    plt.imshow(cv2.cvtColor(sampled_image, cv2.COLOR_BGR2RGB))
    plt.title(f'Sampled Image (Factor: {sampling_factor})')


    plt.subplot(1, 3, 3)
    plt.imshow(quantized_image)
    plt.title(f'Quantized Image (Levels: {quantization_levels})')


    plt.show()


uploaded = files.upload()

image_path = list(uploaded.keys())[0]

# Setting the sampling factor and quantization levels
sampling_factor = 16
quantization_levels = 16

visualize_sampling_quantization(image_path, sampling_factor, quantization_levels)
```

```python
# Variation 1
sampling_factor_1 = 8
quantization_levels_1 = 8
print(f"Sampling Factor: {sampling_factor_1}, Quantization Levels: {quantization_levels_1}")
visualize_sampling_quantization(image_path, sampling_factor_1, quantization_levels_1)

# Variation 2
sampling_factor_2 = 16
quantization_levels_2 = 4
print(f"Sampling Factor: {sampling_factor_2}, Quantization Levels: {quantization_levels_2}")
visualize_sampling_quantization(image_path, sampling_factor_2, quantization_levels_2)

# Variation 3
sampling_factor_3 = 12
quantization_levels_3 = 12
print(f"Sampling Factor: {sampling_factor_3}, Quantization Levels: {quantization_levels_3}")
visualize_sampling_quantization(image_path, sampling_factor_3, quantization_levels_3)

# Variation 4
sampling_factor_4 = 10
quantization_levels_4 = 6
print(f"Sampling Factor: {sampling_factor_4}, Quantization Levels: {quantization_levels_4}")
visualize_sampling_quantization(image_path, sampling_factor_4, quantization_levels_4)

# Variation 5
sampling_factor_5 = 20
quantization_levels_5 = 8
print(f"Sampling Factor: {sampling_factor_5}, Quantization Levels: {quantization_levels_5}")
visualize_sampling_quantization(image_path, sampling_factor_5, quantization_levels_5)
```

```python
import cv2
import numpy as np
from google.colab import files
from matplotlib import pyplot as plt

def linear_contrast(image_rgb, alpha=1.5, beta=100):
    return cv2.convertScaleAbs(image_rgb, alpha=alpha, beta=beta)

def gamma_correction(image_rgb, gamma=0.2):
    gamma_correction = ((image_rgb / 255.0) ** gamma) * 255
    return np.clip(gamma_correction, 0, 255).astype(np.uint8)

def histogram_equalization(image_rgb):
    gray_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
    equalized_image = cv2.equalizeHist(gray_image)
    return cv2.cvtColor(equalized_image, cv2.COLOR_GRAY2RGB)

def enhance_and_display(image_path):

    image = cv2.imread(image_path)

    # Converting the image to RGB (OpenCV uses BGR by default)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Applying different point processing techniques for image enhancement
    enhanced_linear = linear_contrast(image_rgb)
    enhanced_gamma = gamma_correction(image_rgb)
    enhanced_hist_eq = histogram_equalization(image_rgb)

    # Displaying the original and enhanced images using matplotlib
    plt.figure(figsize=(18, 6))

    plt.subplot(1, 4, 1)
    plt.imshow(image_rgb)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 4, 2)
    plt.imshow(enhanced_linear)
    plt.title('Linear Contrast Enhancement')
    plt.axis('off')

    plt.subplot(1, 4, 3)
    plt.imshow(enhanced_gamma)
    plt.title('Gamma Correction')
    plt.axis('off')

    plt.subplot(1, 4, 4)
    plt.imshow(enhanced_hist_eq)
    plt.title('Histogram Equalization')
    plt.axis('off')

    plt.show()

uploaded = files.upload()

image_path = list(uploaded.keys())[0]

enhance_and_display(image_path)
```

```python
import cv2
import numpy as np
from google.colab import files
from matplotlib import pyplot as plt

def linear_contrast(image_rgb, alpha=1.5, beta=100):
    return cv2.convertScaleAbs(image_rgb, alpha=alpha, beta=beta)

def gamma_correction(image_rgb, gamma=0.2):
    gamma_correction = ((image_rgb / 255.0) ** gamma) * 255
    return np.clip(gamma_correction, 0, 255).astype(np.uint8)

def histogram_equalization(image_rgb):
    gray_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
    equalized_image = cv2.equalizeHist(gray_image)
    return cv2.cvtColor(equalized_image, cv2.COLOR_GRAY2RGB)

def enhance_and_display(image_rgb, title):
    plt.imshow(image_rgb)
    plt.title(title)
    plt.axis('off')

uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Loading the original image
original_image = cv2.imread(image_path)
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

# Displaying the images
plt.figure(figsize=(18, 6))

plt.subplot(1, 5, 1)
enhance_and_display(original_image_rgb, 'Original Image')

# Linear Contrast Enhancement with different parameters
enhanced_linear_1 = linear_contrast(original_image_rgb, alpha=1.2, beta=80)
plt.subplot(1, 5, 2)
enhance_and_display(enhanced_linear_1, 'Linear (alpha=1.2, beta=80)')

enhanced_linear_2 = linear_contrast(original_image_rgb, alpha=1.8, beta=50)
plt.subplot(1, 5, 3)
enhance_and_display(enhanced_linear_2, 'Linear (alpha=1.8, beta=50)')

# Gamma Correction with different parameters
enhanced_gamma_1 = gamma_correction(original_image_rgb, gamma=0.5)
plt.subplot(1, 5, 4)
enhance_and_display(enhanced_gamma_1, 'Gamma (gamma=0.5)')

enhanced_gamma_2 = gamma_correction(original_image_rgb, gamma=0.8)
plt.subplot(1, 5, 5)
enhance_and_display(enhanced_gamma_2, 'Gamma (gamma=0.8)')

plt.show()
```

```python
# 3) Mask Processing
```

```python
import cv2
import numpy as np
from google.colab import files
from matplotlib import pyplot as plt

uploaded = files.upload()
image_path = list(uploaded.keys())[0]

image = cv2.imread(image_path)

# Creating two different binary masks
mask1 = np.zeros_like(image, dtype=np.uint8)
cv2.rectangle(mask1, (50, 50), (200, 200), (255, 255, 255), thickness=cv2.FILLED)

mask2 = np.zeros_like(image, dtype=np.uint8)
cv2.circle(mask2, (300, 300), 100, (255, 255, 255), thickness=cv2.FILLED)

# Applying the masks to the image
result1 = cv2.bitwise_and(image, mask1)
result2 = cv2.bitwise_and(image, mask2)

plt.figure(figsize=(18, 5))

plt.subplot(1, 5, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 5, 2)
plt.imshow(mask1[:, :, 0], cmap='gray')
plt.title('Mask 1')
plt.axis('off')

plt.subplot(1, 5, 3)
plt.imshow(cv2.cvtColor(result1, cv2.COLOR_BGR2RGB))
plt.title('Result 1 after Masking')
plt.axis('off')

plt.subplot(1, 5, 4)
plt.imshow(mask2[:, :, 0], cmap='gray')
plt.title('Mask 2')
plt.axis('off')

plt.subplot(1, 5, 5)
plt.imshow(cv2.cvtColor(result2, cv2.COLOR_BGR2RGB))
plt.title('Result 2 after Masking')
plt.axis('off')

plt.show()
```
